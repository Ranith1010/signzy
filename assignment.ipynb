```python
def clean_model_name(model):
    # Convert to lowercase
    model = model.lower()
    # Remove special characters and extra spaces
    model = re.sub(r'[^a-zA-Z0-9\s]', '', model)
    model = model.strip()
    # Replace common variations with simplified names
    if 'apache' in model:
        return 'apache'
    elif 'jupiter' in model:
        return 'jupiter'
    elif 'ntorq' in model:
        return 'ntorq'
    elif 'sport' in model:
        return 'sport'
    elif 'star city' in model:
        return 'star city'
    elif 'xl100' in model:
        return 'xl100'
    elif 'victor' in model:
        return 'victor'
    elif 'radeon' in model:
        return 'radeon'
    elif 'phoenix' in model:
        return 'phoenix'
    elif 'pulsar' in model:
        return 'pulsar'
    elif 'ronin' in model:
        return 'ronin'
    elif 'raider' in model:
        return 'raider'
    elif 'xl' in model:
        return 'xl series'
    elif 'scotty' in model or 'zest' in model:
        return 'scotty'
    elif 'volvo' in model:
        return 'volvo'
    else:
        return 'other'
```

```python
df1['cleaned_model']=df1['model'].apply(clean_model_name)
```

```python
premium_models = ['pulsar', 'apache', 'ntorq', 'jupiter']
mass_market_models = ['star city', 'phoenix', 'xl series', 'raider', 'ronin', 'radeon','sport','victor']
car_model='volvo'

def categorize_model(cleaned_model):
    if cleaned_model in premium_models:
        return "Premium two-wheeler"
    elif cleaned_model in mass_market_models:
        return "Mass_market_two_wheeler"
    elif cleaned_model in car_model:
        return 'Premium_four_wheeler'
    else:
        return "Other category"
```

```python
df1['vehicle_category1']=df1['cleaned_model'].apply(categorize_model)
```

```python
def renew(days):
    if days<=365:
        return 'High'
    elif days>365 and days<1095:
        return 'Medium'
    else:
        return 'Low'
```

```python
df1['insurance_renewal']=df1['days'].apply(renew)
```

```python
def renew_probability(vehicle_ownership, vehicle_category1):
    if vehicle_ownership == 'single_owner':
        if vehicle_category1 == 'Premium two-wheeler' or vehicle_category1 == 'Premium_four_wheeler':
            return 'High'
        else:
            return 'Low'
    elif vehicle_ownership == 'multiple_owners' :
        if vehicle_category1 == 'Mass_market_two_wheeler' or vehicle_category1 == 'Premium two-wheeler' or vehicle_category1 == 'Premium_four_wheeler':
            return 'Medium'
        else:
            return 'Low'
    else:
        return 'Low'  
    
df1['renewal_probability'] = df1.apply(lambda row: renew_probability(row['vehicle_ownership'], row['vehicle_category1']), axis=1)
```
